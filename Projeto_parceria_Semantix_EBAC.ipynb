```python
describe_df = df.describe()
num_cols = len(describe_df.columns)
num_rows = (num_cols + 1) // 2

fig, axes = plt.subplots(num_rows, 2, figsize=(12, 6 * num_rows))

for i, column in enumerate(describe_df.columns):
    row = i // 2  # Calcula a linha atual
    col = i % 2   # Calcula a coluna atual

    ax = axes[row, col] if num_rows > 1 else axes[col]

    sns.histplot(df[column], kde=True, ax=ax)
    ax.set_title(f'Histograma de {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Frequência')

# Oculta os subplots extras, se houver
if num_cols % 2 != 0:
    fig.delaxes(axes[num_rows - 1, 1])

plt.tight_layout()
plt.show()
```

**Observações**

**- Idade (Age):** A distribuição da idade parece ser assimétrica à direita (com um leve enviesamento positivo), o que indica que a maioria dos indivíduos na amostra é relativamente jovem, com alguns indivíduos mais velhos. Isso pode ser relevante, pois a obesidade pode ter diferentes padrões e causas em diferentes faixas etárias.

**- Altura (Height):** A distribuição da altura parece ser aproximadamente normal (formato de sino), com a maioria dos indivíduos concentrados em torno da média. Pode haver uma ligeira diferença entre as alturas de homens e mulheres, o que é esperado.

**- Peso (Weight):** A distribuição do peso parece ser assimétrica à direita, similar à idade, o que indica que a maioria dos indivíduos tem pesos mais próximos da média, com alguns indivíduos com pesos mais elevados. Essa assimetria é um indício da presença de outliers (valores extremos) que podem representar casos de obesidade mais severa.

```python
# Para os boxplots:
fig, axes = plt.subplots(num_rows, 2, figsize=(12, 6 * num_rows))

for i, column in enumerate(describe_df.columns):
    row = i // 2
    col = i % 2

    ax = axes[row, col] if num_rows > 1 else axes[col]

    sns.boxplot(y=df[column], ax=ax)
    ax.set_title(f'Boxplot de {column}')
    ax.set_ylabel(column)

# Oculta os subplots extras, se houver
if num_cols % 2 != 0:
    fig.delaxes(axes[num_rows - 1, 1])

plt.tight_layout()
plt.show()
```

**Observe:**

- Essa variável "FCVC" significa indica se o indivíduo costuma comer vegetais em suas refeições, e provavelmente esse número está relacionada por dia, ou seja, o consumo de vegetais p/dia.
- Entretanto os números estão na sua maioria quebrados, portanto, não faz sentido uma pessoa consumir 2.927409 por exemplo. Desta forma terei que aredondar todos os números, para posteriormente aplicar á um modelo.

**Raciocínio:**

- Consumo de vegetais é discreto: A variável representa a quantidade de vezes que uma pessoa consome vegetais em suas refeições, que naturalmente é um número inteiro. Não se consome 2.3 ou 1.7 porções de vegetais em uma refeição, por exemplo.
- Números quebrados indicam erro: Os valores quebrados em "FCVC" provavelmente representam algum erro na coleta ou processamento dos dados. Poderiam ser resultados de uma média calculada a partir de respostas mais detalhadas ou de uma conversão incorreta de valores.
- Correção é necessária: Para utilizar essa variável em um modelo preditivo, é fundamental corrigir os valores quebrados, convertendo-os para números inteiros. Isso garantirá a consistência dos dados e uma interpretação mais precisa dos resultados.



**E qual é a sugestão?**

- **Arredondamento:** Uma possível solução para corrigir os valores quebrados é arredondá-los para o inteiro mais próximo, utilizando a função round() do Python. Essa abordagem é simples e preserva a essência da variável.

```python
# Arredondando os valores da coluna 'FCVC'
df['FCVC'] = df['FCVC'].round()

# Verificando a nova distribuição dos valores
print(df['FCVC'].value_counts())
```

## 4. Modelagem / análise

**Relembrando o nosso objetivo nesse ponto:**

- Nesse caso específico, nós precisamos usar um algoritmo de classificação porque a variável alvo (Obesity) é uma variável categórica, e o objetivo do  projeto é prever a categoria de obesidade de um indivíduo.

- A variável alvo Obesity representa o nível de obesidade de um indivíduo e possui diferentes categorias, como "Peso Insuficiente", "Peso Normal", "Sobrepeso", "Obesidade Tipo I", etc. Essas categorias são discretas e não representam um valor numérico contínuo.

- O objetivo do projeto é prever a categoria de obesidade de um indivíduo com base em suas características (features), como hábitos alimentares, nível de atividade física e características demográficas. Ou seja, queremos classificar o indivíduo em uma das categorias de obesidade.

- Os algoritmos de classificação são projetados para prever a categoria de uma variável alvo categórica. Eles aprendem com os dados de treinamento a identificar padrões e relações entre as features e as categorias, para que possam classificar novos dados em uma das categorias existentes.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Criando uma instância do modelo de Regressão Logística
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42)

# Treinando o modelo com os dados de treino
model.fit(X_train, y_train)
```

**Aqui está uma análise para interpretar cada item no relatório de classificação:**

**1. Precisão (Precision):**
- **Definição:** A precisão mede a proporção de predições corretas para uma classe específica em relação ao total de predições feitas para aquela classe. Em outras palavras, de todas as vezes que o modelo previu uma determinada classe, quantas vezes ele acertou?
- **Fórmula:**  Precisão = Verdadeiros Positivos / (Verdadeiros Positivos + Falsos Positivos)
- **Exemplo:** Se um modelo previu 10 casos como "Obesidade Tipo I" e 8 deles estavam realmente corretos, a precisão para a classe "Obesidade Tipo I" seria 8/10 = 0.8.

**2. Recall (Revocação):**

- **Definição:** O recall mede a proporção de predições corretas para uma classe específica em relação ao total de casos reais daquela classe. Em outras palavras, de todos os casos reais de uma determinada classe, quantos o modelo conseguiu prever corretamente?

- **Fórmula:** Recall = Verdadeiros Positivos / (Verdadeiros Positivos + Falsos Negativos)

- **Exemplo:** Se houver 15 casos reais de "Obesidade Tipo I" no conjunto de dados e o modelo previu corretamente 10 deles, o recall para a classe "Obesidade Tipo I" seria 10/15 = 0.67.

**3. F1-score (Pontuação F1):**

- **Definição:** O F1-score é uma média harmônica entre precisão e recall. Ele fornece uma medida balanceada do desempenho do modelo, considerando tanto a precisão quanto o recall.

- **Fórmula:** F1-score = 2 * (Precisão * Recall) / (Precisão + Recall)

- **Exemplo:** Se a precisão para a classe "Obesidade Tipo I" for 0.8 e o recall for 0.67, o F1-score seria 2 * (0.8 * 0.67) / (0.8 + 0.67) = 0.73.

**4. Suporte (Support):**

- **Definição:** O suporte é o número de casos reais de uma determinada classe no conjunto de dados.

- **Exemplo:** Se houver 15 casos reais de "Obesidade Tipo I" no conjunto de dados, o suporte para a classe "Obesidade Tipo I" seria 15.

**5. Macro Average (Média Macro):**

- **Definição:** A média macro é a média aritmética das pontuações de precisão, recall e F1-score para todas as classes, sem levar em conta o suporte. Ela é útil para avaliar o desempenho geral do modelo em todas as classes, independentemente do seu tamanho.

**6. Weighted Average (Média Ponderada):**

- **Definição:** A média ponderada é a média das pontuações de precisão, recall e F1-score para todas as classes, ponderada pelo suporte de cada classe. Ela é útil para avaliar o desempenho geral do modelo, levando em conta o tamanho de cada classe.

**7. Accuracy (Acurácia):**

- **Definição:** A acurácia é a proporção de predições corretas em relação ao total de predições feitas. Ela mede o desempenho geral do modelo em todas as classes.

- **Fórmula:** Acurácia = (Verdadeiros Positivos + Verdadeiros Negativos) / (Verdadeiros Positivos + Verdadeiros Negativos + Falsos Positivos + Falsos Negativos)

**Considerações:**

- **Precisão vs. Recall:** É importante equilibrar a precisão e o recall, dependendo do problema. Em alguns casos, é mais importante ter alta precisão, mesmo que isso signifique um recall menor (por exemplo, em diagnósticos médicos). Em outros casos, é mais importante ter alto recall, mesmo que isso signifique uma precisão menor (por exemplo, em detecção de fraudes).

- **F1-score:** O F1-score é uma boa métrica para usar quando você precisa de um equilíbrio entre precisão e recall.

- **Suporte:** O suporte é importante para entender a representatividade de cada classe no conjunto de dados. Classes com baixo suporte podem ter pontuações menos confiáveis.

- **Médias:** As médias macro e ponderada fornecem uma visão geral do desempenho do modelo em todas as classes.

## 4.2 Curva ROC e AUC

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.preprocessing import label_binarize
import numpy as np

# Binarizar a saída
y_test_bin = label_binarize(y_test, classes=np.unique(y_test))
n_classes = y_test_bin.shape[1]

# Calcular probabilidades para cada classe
y_prob = model.predict_proba(X_test)

# Calcular a curva ROC e a área ROC para cada classe
fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# CCalcular a curva ROC micromédia e a área ROC
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_prob.ravel())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Plotar curvas ROC
plt.figure(figsize=(10, 8))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'ROC curve of class {np.unique(y_test)[i]} (area = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de falsos positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.title('Curvas ROC para classificação multiclasse')
plt.legend(loc="lower right")
plt.show()


# Calcular AUC para a micromédia
auc_micro = roc_auc_score(y_test_bin, y_prob, average='micro')
print(f'AUC micromédia: {auc_micro:.2f}')

# Calcular AUC para a macro-média
auc_macro = roc_auc_score(y_test_bin, y_prob, average='macro')
print(f'Macro-média AUC: {auc_macro:.2f}')
```

**Observações sobre o resultado da Curva ROC e AUC**

- Com base nesses resultados, podemos concluir que o modelo de classificação tem um excelente desempenho na tarefa de prever a categoria de obesidade de um indivíduo, tanto globalmente quanto individualmente para cada classe. Os valores altos da AUC micromédia e macro-média indicam que o modelo tem uma alta capacidade de discriminação entre as classes e uma alta probabilidade de classificar corretamente os exemplos positivos.

- O modelo apresenta uma precisão alta e consistente em todas as classes, o que o torna uma ferramenta confiável para a previsão do nível de obesidade.
